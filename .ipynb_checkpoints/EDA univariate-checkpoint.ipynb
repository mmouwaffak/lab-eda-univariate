### Part 2: Delving into Product Pricing

**Business Question**: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `price` of products.
    - What's the average price point of products listed? How does this compare with the most common price point (mode)?

The average price of products listed is 89.24, but the most common price is 9.99. It means that most products are low priced, and that the expensive products pull the average up.

```python
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]         # [0] gets the first mode value

print(f"Mean Price: {mean_price:.2f}")            # UK : pounds 
print(f"Median Price: {median_price:.2f}")     # price: rounded 2 decimals
print(f"Mode Price: {mode_price:.2f}")
```

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, range, and interquartile range for product `price`.
    - How varied are the product prices? Are there any indicators of a significant spread in prices?

Product prices are varied (range of 100000). The IQR is at 36.00, meaning that 50% of the products have a price below or equal to 36.
Most products are low priced, but there are luxury items, creating a huge price variation.

```python
# Variance, standard deviation, range, interquatile (IQR)


variance_price = df['price'].var()
std_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()

q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr_price = q3 - q1


print(f"Variance: {variance_price:.2f}")
print(f"Standard Deviation: {std_price:.2f}")
print(f"Range: {range_price:.2f}")
print(f"IQR: {iqr_price:.2f}")
```

3. **Visualizations**:
    - Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. *If its hard to read these diagrams, think why this is, and explain how it could be solved.*.
    - Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing. 

The majority of the product has a price range below 2000 ; it is hard the read because of the extreme values of the outliers. 
We can observe 2 outliers above 25000 : one at 80000 and one at 100000 (the extremity of the price range.

```python
# Histogram



plt.figure(figsize=(10, 6))
sns.histplot(df['price'])
plt.title('Distribution of Product Prices')
plt.xlabel('Price ()')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Boxplot

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Product Price Spread and Outliers')
plt.xlabel('Price ()')
plt.tight_layout()
plt.show()
```

### Part 3: Unpacking Product Ratings

**Business Question**: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

1. **Measures of Centrality**:
    - Calculate the mean, median, and mode for the `rating` of products.
    - How do customers generally rate products? Is there a common trend?

The customers generally evaluate with `stars`. Most products on Amazon have no rating, and the average is 2.15.

```python
# Mean, median, mode for rating

mean_rating = df['stars'].mean()            #rating
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print(f"Mean Rating: {mean_rating:.2f}")
print(f"Median Rating: {median_rating:.2f}")
print(f"Mode Rating: {mode_rating:.2f}")
```

2. **Measures of Dispersion**:
    - Determine the variance, standard deviation, and interquartile range for product `rating`.
    - Are the ratings consistent, or is there a wide variation in customer feedback?

```python
# Variance, standard deviation, IQR (interquatile)

variance_rating = df['stars'].var()
std_rating = df['stars'].std()
q1_rating = df['stars'].quantile(0.25)
q3_rating = df['stars'].quantile(0.75)
iqr_rating = q3_rating - q1_rating

print(f"Variance: {variance_rating:.2f}")
print(f"Standard Deviation: {std_rating:.2f}")
print(f"IQR: {iqr_rating:.2f}")
```

3. **Shape of the Distribution**:
    - Calculate the skewness and kurtosis for the `rating` column. 
    - Are the ratings normally distributed, or do they lean towards higher or lower values?

```python
#  Skewness and kurtosis

from scipy import stats

skewness = df['stars'].skew()
kurtosis = df['stars'].kurtosis()

print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

4. **Visualizations**:
    - Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['stars'])
plt.title('Distribution of Product Ratings')
plt.xlabel('Stars (Rating)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
